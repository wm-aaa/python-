```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

headers = { 
        "Referer":"https://coding.imooc.com/?c=java",
        "Sec-Fetch-Dest":"script",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36",
}
url = "https://coding.imooc.com/?c=miniprogram"
response= requests.get(url,headers=headers)   #get请求
html = response.text
mysout = BeautifulSoup(html,"lxml")    #转换为lxml
Cname=[]
Cimg=[]
Curl=[]
cintroduce=[]
div_1 = mysout.find('div',class_='w index-main')
div_2 = div_1.find('div',class_='index-list-wrap')
div_3 = div_2.find('div',class_='shizhan-course-list clearfix')
div_all = div_3.find_all('div',class_='shizhan-course-wrap l')
for div in div_all[1:]:
    a=div.find('a')
    Curl.append("https://coding.imooc.com/"+a.get('href'))
    div=div.find('div',class_='box')
    img=div.find('img')
    Cimg.append("https:"+img.get('src'))
    div=div.find('div',class_='shizhan-intro-box')
    p = div.find('p',class_="shizan-name")
    Cname.append(p.get('title'))
    p = div.find('p',class_="shizan-desc")
    cintroduce.append(p.get('title'))
```

```python
df = pd.DataFrame(Cname,columns=['Cname'])
df['Cimg']=Cimg
df['Curl']=Curl
df['cintroduce']=cintroduce
df
```

```python
df.to_csv('E:\python_data\前端开发\小程序.csv',encoding='utf-8-sig',index=False)
```
